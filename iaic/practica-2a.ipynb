```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
```

La formula: "Y = 0.7*x + 60 + <número aleatorio entre -5.9 y +5.9>", propuesta en la práctica, siempre va a tender a incrementar debido a que el valor de la pendiente "0,7 * x" siempre va a ser positiva. 

El valor de 60 indica el punto de partida, es decir, la frecuencia inicial. Se escoge 60 BPM ya que es la frecuencia mínima la cual un adulto promedio tiene en un estado de reposo. 

```python
def generarDatosFC():
    x = np.arange(1, 61)
    y = 0.7 * x + 60 + np.random.uniform(low=-5.9, high=5.9, size=x.shape)
    return x, y
```

```python
x, y = generarDatosFC()
```

```python
X = [[i] for i in x]
```

```python
reg = LinearRegression()
reg.fit(X, y)
print(f"Coefficients: {reg.coef_} Intercept: {reg.intercept_}")
```

```python
yLinealResult = reg.predict(X)
```

```python
plt.scatter(x, y)
plt.plot(x, yLinealResult)
plt.show()
```

El coeficiente e intercepto reales son de: 0,7 y 60, respectivamente. El coeficiente e intercepto entrenados son de: 0.72653941 y 59.33712168564188, respectivamente.

La conclusión que se obtiene es:
- El modelo de regresión lineal, en este caso, es preciso debido a que los datos obtenidos son casi identicos a los reales, indicando una captación correcta de la tendencia de los datos.
- El impacto del ruido aleatorio definido es mínimo, indicando que la frecuencia cardiaca sigue un patron lineal estable.

**Coenficiente de determinación: score(X, y)**

Gracias a este atributo, se puede evaluar la precisión del modelo. Mientras el resultado esté más cercano al 1, signfica que la regresión esta calculando correctamente la relación entre ambas variables. En nuestro caso, el resultado es extremadamente cercano a 1, por lo tanto confirma la fuerte relación existente entre las dos variables.

**Predicción: predict(X)**

Este método se utiliza para predecir los valores de salida (y), a partir de nuevos datos de la variable de entrada (X), utilizando el modelo previamente entrenado. Con este método se puede predecir que frecuencia cardiaca en un segundo fuera del rango preestablecido (en nuestro caso, fuera de 1 y 60). En el ejemplo propuesto, en el segundo 90, la frecuencia es de ~124 BPM, concluyendo que el usuario podría estas realizando una actividad física de fuerza moderada.

```python
rel = reg.score(X, y)
pred = reg.predict([[90]])
print(f"Relacion: {rel}")
print(f"Predicción: {pred}")
```
