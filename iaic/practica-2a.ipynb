```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
def generarDatosFC():
    x = np.arange(1, 61)
    y = 0.7 * x + 60 + np.random.uniform(low=-5.9, high=5.9, size=x.shape)
    return x, y
```

```python
x, y = generarDatosFC()
```

```python
X = [[i] for i in x]
```

```python
reg = LinearRegression()
reg.fit(X, y)
print(f"Coefficients: {reg.coef_} Intercept: {reg.intercept_}")
```

```python
yLinealResult = reg.predict(X)
```

```python
plt.scatter(x, y)
plt.plot(x, yLinealResult)
plt.show()
```
